```python
finaldf=pd.merge(df,dfC,on='Country Code',how='left')
```

```python
finaldf.Country.value_counts()
```

```python
country_names=finaldf.Country.value_counts().index
```

```python
country_val=finaldf.Country.value_counts().values
```

```python
# PIE chart : top 3 countries to use zomato 
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

As we can see from the chart that Maximum of the zomato records and transactions are from India and then USA and UK 

```python
finaldf.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=finaldf.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

# Observations
1.When ratings are from  4.5 to 4.9----> Excellant 
2.When ratings are from  4.0 to 4.4----> very good
3.When ratings are from  3.5 to 3.9----> Good
4.When ratings are from  3.0 to 3.4----> average
5.When ratings are from  2.5 to 2.9----> average
6.When ratings are from  2.0 to 2.4----> poor

```python
sns.set({'figure.figsize':(10,5)})
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

As we can see frok thr graph that the most of the ratings are from the no rating bar and after that 3.2 ratings by 500 rating count

```python
# Adding colour palette
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['White','Red','Orange','Yellow','Green','Green'])
```

Observations :
not rated is very high 
most of the ratings are from 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['Pink','Red','Orange','Yellow','Green','Green'])
```
